Importing Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Loading DataFrame and making updates

```python
df = pd.DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/jward6301/datasci_3_eda/main/datasets/healthcarediabetes.csv')
df
```

```python
df.size
df.shape
```

```python
df.sample(10)
```

```python
#removing duplicate rows
df.drop_duplicates()
df
```

```python
#removing missing values
df.dropna()
df
```

```python
#Replacing unwanted characters in column names such as spaces
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace(r'[^a-zA-z0-9]', '')
df.columns = df.columns.str.lower()
df
```

```python
#extra line to furhter clean up column names
df.columns = df.columns.str.strip()
df
```

```python
#Measure of spread

range_pregnancies = df['pregnancies'].max() - df['pregnancies'].min()
var_pregnancies = np.var(df['pregnancies'])
std_pregnancies = np.std(df['pregnancies'])
percentile_25_pregnancies = np.percentile(df['pregnancies'], 25)
percentile_75_pregnancies = np.percentile(df['pregnancies'], 75)
q1_pregnancies = np.percentile(df['pregnancies'], 25)
q3_pregnancies = np.percentile(df['pregnancies'], 75)
iqr_pregnancies = q3_pregnancies - q1_pregnancies

range_bmi = df['bmi'].max() - df['bmi'].min()
var_bmi = np.var(df['bmi'])
std_bmi = np.std(df['bmi'])
percentile_25_bmi = np.percentile(df['bmi'], 25)
percentile_75_bmi = np.percentile(df['bmi'], 75)
q1_bmi = np.percentile(df['bmi'], 25)
q3_bmi = np.percentile(df['bmi'], 75)
iqr_bmi = q3_bmi - q1_bmi
```

```python
print('Range for pregnancies:', range_pregnancies)
print('Variance for pregnancies:', var_pregnancies)
print('Standard deviation for pregnancies:', std_pregnancies)
print('25th percentile for pregnancies:', percentile_25_pregnancies)
print('75th percentile for pregnancies:', percentile_75_pregnancies)
print('IQR for pregnancies:', iqr_pregnancies)
print('Range for bmi:', range_bmi)
print('Variance for bmi:', var_bmi)
print('Standard deviation for bmi:', std_bmi)
print('25th percentile for bmi:', percentile_25_bmi)
print('75th percentile for bmi:', percentile_75_bmi)
print('IQR for bmi:', iqr_bmi)
```

```python
# histogram for pregnancies

mean = 3.7427745664739884
std_dev = 3.323801
num_samples = 1000


plt.hist(df['pregnancies'], bins=30, density=True, alpha=0.7, color='blue', label='Histogram')


x = np.linspace(min(df['pregnancies']), max(df['pregnancies']), 100)
pdf = (1/(std_dev * np.sqrt(2*np.pi))) * np.exp(-(x - mean)**2 / (2*std_dev**2))
plt.plot(x, pdf, color='green', label='PDF')

plt.title('Patient Pregnancies')
plt.xlabel('Average Pregnancies')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# histogram for bmi

mean = 32.137392
std_dev = 8.076127
num_samples = 1000

plt.hist(df['bmi'], bins=30, density=True, alpha=0.7, color='blue', label='Histogram')

x = np.linspace(min(df['bmi']), max(df['bmi']), 100)
pdf = (1/(std_dev * np.sqrt(2*np.pi))) * np.exp(-(x - mean)**2 / (2*std_dev**2))
plt.plot(x, pdf, color='green', label='PDF')

plt.title('Patient BMI')
plt.xlabel('Average BMI')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

Handling Outliers:

```python

q1 = np.percentile(df['pregnancies'], 25)
q3 = np.percentile(df['pregnancies'], 75)
iqr = q3 - q1

threshold_lower = q1 - 1.5 * iqr
threshold_upper = q3 + 1.5 * iqr

outliers = [x for x in df['pregnancies'] if x < threshold_lower or x > threshold_upper]

# Printing outliers

outliers_list = outliers
outliers_list.sort()
print(outliers_list)
```

Handling outliers

* All 16 outliers will be included as it may be necessary in determing aspects of the patients health and their medical records regarding their pregnancies. It seems as if these entries are correct as there are a few with these numbers but they seem to be uncommon numbers of pregnancies
